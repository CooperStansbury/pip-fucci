```python
import skimage
import pandas as pd
import numpy as np
from tifffile import imread
import matplotlib.pyplot as plt
```

```python
fpath = "/scratch/indikar_root/indikar1/cstansbu/pf-test/images/test.processed.tiff"

img = imread(fpath)
print(f"{img.shape=}")
```

```python
t = 250

plt.rcParams['figure.dpi'] = 300

fig, axs = plt.subplots(1, 3)

axs[0].imshow(img[t, 0, :, :], cmap='Reds')
axs[1].imshow(img[t, 1, :, :], cmap='Blues')
axs[2].imshow(img[t, 2, :, :], cmap='Greens')

axs[0].axis(False)
axs[1].axis(False)
axs[2].axis(False)

plt.tight_layout()
```
