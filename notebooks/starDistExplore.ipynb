```python
def data2prob(data):
    """a function to return the label to probability mapping"""
    probs = data['prob']
    newRows = []
    for i in range(1, len(probs) + 1):
        row = {
            'label' : i,
            'prob' : probs[i-1],
        }
        newRows.append(row)
    return pd.DataFrame(newRows)


px = data2prob(data)     
px.head()
```
