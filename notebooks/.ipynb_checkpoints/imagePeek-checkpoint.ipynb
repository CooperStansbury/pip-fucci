```python
import pandas as np
import numpy as np
import matplotlib.pyplot as plt
from tifffile import  imread
```

```python
rawpath = "/scratch/indikar_root/indikar1/cstansbu/pf-test/images/A3.raw.tiff"
imgpath = "/scratch/indikar_root/indikar1/cstansbu/pf-test/images/A3.processed.tiff"
segpath = "/scratch/indikar_root/indikar1/cstansbu/pf-test/images/A3.segmented.tiff"

t=0
raw = imread(rawpath, key=t)
img = imread(imgpath, key=t)
seg = imread(segpath, key=t)
print(f"{raw.shape=} {img.shape=} {seg.shape=}")
```

```python
c = 0

plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = 10, 3
fig, axs = plt.subplots(1, 3)
axs = axs.ravel()

axs[0].imshow(raw[c, :, :])
axs[0].axis(False)
axs[1].imshow(img[c, :, :])
axs[1].axis(False)
axs[2].imshow(seg)
axs[2].axis(False)

plt.tight_layout()
```
